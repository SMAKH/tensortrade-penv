```python
import torch
import matplotlib.pyplot as plt

concentration=torch.tensor([1,5],dtype=torch.float)
dist = torch.distributions.dirichlet.Dirichlet(concentration=concentration, validate_args=True)
x=[]
y=[]
for i in range(100):
    s=dist.sample()
    x.append(s[0])
    y.append(s[1])
plt.scatter(x,y)
plt.plot()
```

```python
print(dist.log_prob(torch.tensor([.1,.9],dtype=torch.float32)))
```
