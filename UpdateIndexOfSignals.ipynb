```python
import pickle
with open("./boors/new_symbol_index.pck",'rb') as ff:
    symbol_maps=pickle.load(ff)
```

```python
import pandas as pd
signals=pd.read_csv("./boors/important_signals.csv")
signals.head()
```

```python
imp_experts=pd.read_csv("./boors/important_experts.csv")
imp_experts.head()
```

```python
for index,row in signals.iterrows():
  signals.at[index,'symbol_id']=symbol_maps[int(signals.iloc[index]['symbol_id'])]
  signals.at[index,'expert_id']=imp_experts[imp_experts['expert_id']==signals.iloc[index]['expert_id']]['new_id']
```

```python
signals.head()
```

```python
signals.to_csv("./boors/reindexed_important_signals.csv")
```
