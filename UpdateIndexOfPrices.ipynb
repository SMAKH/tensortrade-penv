```python
import pandas as pd
%matplotlib inline
data=pd.read_csv("./boors/main_final_data.csv")
data.head()
```

```python
imp_symbols=pd.read_csv("./boors/important_symbols.csv")
imp_symbols.head()
cols=["date"]
new_index={}
last_index=0
for i in imp_symbols['symbol_id']:
    cols.append(str(i)+":open")
    cols.append(str(i)+":high")
    cols.append(str(i)+":close")
    cols.append(str(i)+":low")
    cols.append(str(i)+":volume")
    new_index[i]=last_index
    last_index+=1
    
```

```python
import pickle
with open("./boors/new_symbol_index.pck",'wb') as file:
    pickle.dump(new_index,file)
```

```python
filtered_data=data[cols]
filtered_data.head()
```

```python
new_index[10]
```

```python
new_columns=['date']
columns=filtered_data.columns[1:]
for c in columns:
    id=int(c.split(":")[0])
    t=c.split(":")[1]
    new_columns.append(str(new_index[id])+":"+t)
```

```python
filtered_data.columns=new_columns
```

```python
filtered_data.to_csv("./boors/reindexed_price_data.csv")
```
