```python
import pandas as pd
df=pd.read_csv("./signals.csv")
df['start_date']=pd.to_datetime(df['start_date'])
df['start_date']=pd.to_datetime(df['close_date'])
df=df[(df['start_date'].notnull()&df['close_date'].notnull())]
```

```python
from matplotlib.pyplot import figure

figure(figsize=(20, 20), dpi=200)
import matplotlib.pyplot as plt
for index,row in df.iterrows():
    dates=pd.date_range(row['start_date'],row['close_date'])
    data=[index]*len(dates)
    plt.plot(dates,data)
    
```

```python
start_date="3/21/2018"
end_date="3/20/2019"
signals=df[(df['start_date']>=start_date)&(end_date>=df['start_date'])]
signals.shape
```
